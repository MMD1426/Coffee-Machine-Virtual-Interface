```python
import cv2
import os
from cvzone.HandTrackingModule import HandDetector
from datetime import datetime
import csv
```

```python
# importing all the Modes to a list
folderPathModes = 'Modes'
listimgModesPath = os.listdir(folderPathModes)
listimgModes = []
for imgModesPath in listimgModesPath:
    listimgModes.append(cv2.imread(os.path.join(folderPathModes,imgModesPath)))

# importing all the Icons to a list
folderPathIcon = 'Icons'
listimgIconPath = os.listdir(folderPathIcon)
listimgIcon = []
for imgIconPath in listimgIconPath:
    listimgIcon.append(cv2.imread(os.path.join(folderPathIcon,imgIconPath)))
```

```python
listName = []
for name in listimgIconPath:
    name = name.replace(name[:2],'')
    name = name.replace('.png','')
    listName.append(name)
print(listName)
```

```python
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_XI_WIDTH,640)
cap.set(cv2.CAP_PROP_XI_HEIGHT,480)

detector = HandDetector()

modesType = 0
selection = -1
counter = 0
modesPositions = [(1136,196),(1000,384),(1136,581)]
selectionList = [-1,-1,-1]
counterPause = 0
counterSpeed = 20

imgbackgraound = cv2.imread('Background.png')

while True:
    ret,frame = cap.read()

    imgbackgraound[139:480+139,50:50+640] = frame
    imgbackgraound[0:720,847:1280] = listimgModes[modesType]

    hands, frame = detector.findHands(frame, draw=True, flipType=True)


    if hands and counterPause == 0 and modesType < 3:
        hand1 = hands[0]
        fingers1 = detector.fingersUp(hand1)
        print(fingers1)

        if (fingers1 == [0,1,0,0,0]) or (fingers1 == [1,1,0,0,0]):
        
            if selection != 1:
                counter = 1
            selection = 1
        elif (fingers1 == [0,1,1,0,0]) or (fingers1 == [1,1,1,0,0]):
        
            if selection != 2:
                counter = 1
            selection = 2
        elif (fingers1 == [0,1,1,1,0]) or (fingers1 == [1,1,1,1,0]):
        
            if selection != 3:
                counter = 1
            selection = 3  

        else:
            selection = -1
            counter = 0 
        
        if counter > 0:
            counter += 1

            cv2.ellipse(imgbackgraound,modesPositions[selection-1],
                        (103,103),0,0,counter*counterSpeed,(0,255,0),20)

        if counter*counterSpeed > 360:
            selectionList[modesType] = selection
            modesType += 1
            counter = 0
            selection = -1
            counterPause = 1


    if counterPause > 0:
        counterPause += 1
        if counterPause >= 15:
            counterPause =0

    if selectionList[0] != -1:
        imgbackgraound[636:636+65,133:133+65] = listimgIcon[selectionList[0] - 1]
    if selectionList[1] != -1:
        imgbackgraound[636:636+65,340:340+65] = listimgIcon[selectionList[1] + 2]
    if selectionList[2] != -1:
        imgbackgraound[636:636+65,542:542+65] = listimgIcon[selectionList[2] + 5]

    
    if modesType == 3:
        counterPause += 1
        if counterPause > 10:
            dateNow = datetime.now().strftime("%Y/%m/%d")
            timeNow = datetime.now().strftime("%H:%M:%S")
            with open('DataBace.csv', 'a', newline='') as f:      
                writer = csv.writer(f, delimiter=',')
                fields = [dateNow,
                        timeNow,
                        listName[selectionList[0] - 1],
                        listName[selectionList[1] + 2],
                        listName[selectionList[2] + 5]]
                writer.writerow(fields)
                f.close()
            break
        

    cv2.imshow("Coffee Machine Virtual Interface",imgbackgraound)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
